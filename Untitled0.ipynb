```
from google.colab import files
import pandas as pd
import io

  
def upload_file():
  uploaded = files.upload()
  return uploaded
  
def main():
  uploaded = upload_file()
  #print(uploaded['BNK48.csv'].decode('utf-8') + '...')
  data = uploaded['BNK48.csv'].decode('utf-8') 
  df = pd.read_csv(io.StringIO(data))
  print(df)
  
if __name__ == "__main__":
  main()
```
